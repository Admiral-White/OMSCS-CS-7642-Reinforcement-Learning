## This notebook is aimed to replicate the random walk experiment in Sutton 1988. 

<img src='./random_walk.jpg' width=800>   
Initial state: **D**    
Terminal states: __A (z=0), G (z=1)__   
If the walk DCDEFG occurred, then the learning procedure would be given the sequence $X_D$, $X_C$, $X_D$, $X_E$, $X_F$, 1

```python
# import necessary modules
import numpy as np
import random as rd
import matplotlib.pyplot as plt
%pylab inline 
import seaborn as sns
```

```python
# create a class for generating training sets
class random_walk:
    def __init__(self,num_states = 7,initial_state = 3,num_sequence=10,num_training_set=100):
        self.num_states = num_states
        self.initial_state = initial_state
        self.num_sequence = num_sequence
        self.num_training_set = num_training_set 
        self.train = self.training_set_generator(self.num_states,self.initial_state,self.num_sequence,self.num_training_set)
        self.ideal_ws = [1/6,1/3,1/2,2/3,5/6]
    def training_set_generator(self,num_states,initial_state,num_sequence,num_training_set):
        np.random.seed(27)
        train = [[None] * num_training_set] * num_sequence
        for i in range(num_sequence):
            for j in range(num_training_set):
                x_vectors = np.zeros(num_states)
                step_now = initial_state
                x_vectors[step_now] = 1 #initial step, in this paper it's 3 aka state D
                while step_now not in [0,num_states-1]:
                    x_vector = np.zeros(num_states)
                    if np.random.uniform() > 0.5: #randomly take a left step or right step (same probability)
                        step_now += 1 
                    else:
                        step_now += -1
                    x_vector[step_now] = 1
                    x_vectors = np.vstack([x_vectors,x_vector])
                # if step_now is in the absorbing state then while loop ends
                train[i][j] = x_vectors
        return train
    def history_p_sum(self,i,lambda_,initial_weights,train_sample_):
        t = i+1 
        lambda_all = [lambda_**(t-j) for j in range(1,t+1)]
        x_all = train_sample_[:t]
        return np.sum(np.array(lambda_all)[:,None] * np.array(x_all),0)
    def update_weights(self,train_sample,alpha=0.01,lambda_=0.3,initial_weights=[0.5,0.5,0.5,0.5,0.5]):
        z = [0,1][int(train_sample[-1][-1])] # the final outcome
        num_nonterm_states = self.num_states - 2
        num_samples = train_sample.shape[0] - 1
        train_sample_ = train_sample[:-1,1:-1] # keep only the non_terminal states 
        delta_w = np.zeros(num_nonterm_states)
        for i in range(num_samples):
            if i == num_samples - 1: # when S_t+1 is the terminal state
                delta_p = z - initial_weights[int(np.argmax(train_sample_[i]))] # the prediction Pt = w_T xt was simply the value of the ith comp onent of w
            else:
                delta_p = initial_weights[int(np.argmax(train_sample_[i+1]))] - initial_weights[int(np.argmax(train_sample_[i]))]
            #print(self.history_p_sum(i,lambda_,initial_weights,train_sample_))
            delta_w += alpha * delta_p * self.history_p_sum(i,lambda_,initial_weights,train_sample_)
        return delta_w
    def train_repeated(self,train,alpha=0.01,lambda_=0.3,initial_weights=[0.5,0.5,0.5,0.5,0.5],num_epochs=20):
        for i in range(num_epochs):
            for j in range(self.num_sequence):
                delta_ws = np.zeros(5)
                for z in range(self.num_training_set):
                    delta_ws += self.update_weights(train[j][z],alpha,lambda_,initial_weights)
                #print(delta_ws)
                initial_weights += delta_ws # update the weights after every sequence
        RMSE = (np.sum((self.ideal_ws - initial_weights)**2)/(self.num_states-2))**0.5
        return RMSE
    def train_single_seq(self,train,alpha=0.01,lambda_=0.3,initial_weights=[0.5,0.5,0.5,0.5,0.5]):
        RMSE = 0
        for j in range(self.num_sequence):
            initial_weights = [0.5,0.5,0.5,0.5,0.5]
            for z in range(self.num_training_set):
                initial_weights += self.update_weights(train[j][z],alpha,lambda_,initial_weights) # update the weights after every training set
                RMSE += (np.sum((self.ideal_ws - initial_weights)**2)/(self.num_states-2))**0.5
        return RMSE/(self.num_sequence*self.num_training_set)
```

```python
RW = random_walk()
```

```python
train = RW.train
```

## Fig. 3

```python
lambdas_ = np.linspace(0,1,20)
RMSEs = []
for i in range(len(lambdas_)):
    lambda_ = lambdas_[i]
    RMSEs.append(RW.train_repeated(train,alpha=0.001,lambda_=lambda_))
    if i % 10 == 0:
        print("{}/{} finished".format(i,len(lambdas_)))
```

```python
plt.figure(figsize=(8,6))
plt.plot(lambdas_,RMSEs,'-o')
plt.ylabel('ERROR',size=20)
plt.xlabel('λ',size=20)
plt.title('Replication of Figure 3 in Sutton 1988',size=15)
plt.annotate('Widrow-Hoff',(0.65, RMSEs[-2]),fontsize=16)
plt.show()
```

## Fig. 4

```python
RW = random_walk(num_sequence=100,num_training_set=10)
train = RW.train
```

```python
lambdas_ = [1,0,0.8,0.3]
alphas = np.linspace(0,0.7,20)
RMSEs = {1:[],0:[],0.8:[],0.3:[]}
for lambda_ in lambdas_:
    print('lambda = {}:'.format(lambda_))
    for i in range(len(alphas)):
        alpha = alphas[i]
        RMSEs[lambda_].append(RW.train_single_seq(train,alpha=alpha,lambda_=lambda_))
        if i % 5 == 0:
            print("{}/{} finished".format(i,len(alphas)))
```

```python
plt.figure(figsize=(10,6))
for i in lambdas_:
    plt.plot(alphas,RMSEs[i],'-o',label='λ = {}'.format(i))
    if i == 1:
        plt.annotate('λ = {} (Widrow-Hoff)'.format(i),(0.64,RMSEs[i][-4]))
    else:    
        plt.annotate('λ = {}'.format(i),(0.72,RMSEs[i][-1]))
plt.xlim([-0.05,0.8])
plt.xticks(np.linspace(0,0.7,8))
plt.ylabel('ERROR',size=20)
plt.xlabel('α',size=20)
plt.title('Replication of Figure 4 in Sutton 1988',size=15)
plt.legend()
plt.show()
```
